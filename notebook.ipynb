```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/athlete_events.csv', delimiter=',', on_bad_lines='skip', encoding='utf-8', engine='python')
```

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(12,6))
top_sports=df['Sport'].value_counts().head(10)
sns.barplot(x=top_sports.values,y=top_sports.index,palette='Set2')
plt.title('Top 10 Sports by Number of Athletes')
plt.xlabel('Number of Athletes')
plt.ylabel('Sport')
plt.tight_layout()
plt.show()
```
